### 2. Membuat dan Mengakses Elemen Series

#### a. Membuat series dari sebuah List Python

```python
import pandas as pd

li = [1, 7, 2]
se = pd.Series(li)
print(se)
```

#### b. Mengakses Elemen pada series

```python
li = [1, 7, 2]
se = pd.Series(li)

print(se[0])
```

#### c. Memberi Label pada Series

```python
li = [1, 7, 2]
se = pd.Series(li, index = ["x", "y", "z"])

print(se['x'])
```

#### d. Membuat series dari Dictionary

```python
calories = {"day1": 420, "day2": 380, "day3": 390}
se = pd.Series(calories)

print(se)
```

### 3. Membuat DataFrame

#### a. Membuat DataFrame

```python
import pandas as pd

df = pd.DataFrame(data={"Nama": ["Ali", "Joko", "Adi"],
                        "Umur": [12, 13, 15],
                        "Kelas": [6, 7, 8]})

df.columns
df.columns.values[0] = "Nama singkat"
print(df)

data = {
  "calories": [420, 380, 390],
  "duration": [50, 40, 45]
}

df = pd.DataFrame(data)
df
```

#### b. Mengakses elemen pada DataFrame

```python
df.iloc[0:2, [0,1]]
df.iloc[0:2, 0:2]
df.iloc[:, 0:2]
df.loc[0:2, :'calories']
df.loc[0:2, ['calories', 'duration']]
df['calories']
df['calories'][0]
df[['calories']]
df[['calories']].loc[[0, 1]]
```

#### c. Memberi nama pada index

```python
data = {
    "calories": [420, 380, 390],
  "duration": [50, 40, 45]
}

df = pd.DataFrame(data, index = ["day1", "day2", "day3"])
print(df)
df
```
