### Perbandingan tipe ruangan AirBnb

```python
# Membuat dataframe untuk menghitung perbandingan tipe ruangan
room_count_df = df.groupby('room_type').count().reset_index()
room_count_df = room_count_df[['room_type', 'neighbourhood_group']]


plt.figure(figsize=(8, 8))
ax = room_count_df.plot(
    kind='pie',
    y='neighbourhood_group',
    labels=room_count_df['room_type'],
    autopct='%.2f%%',
    pctdistance=0.85,
    labeldistance=1.05,
    textprops={'fontsize': 10}
)

# Judul dan legenda
plt.title(
    label="Rasio tipe ruangan AirBnb di New York City",
    fontdict={"fontsize": 16}
)
plt.legend(
    loc='lower right',
    bbox_to_anchor=(1.3, 0)
)


plt.ylabel('')
plt.tight_layout()
plt.show()
```
