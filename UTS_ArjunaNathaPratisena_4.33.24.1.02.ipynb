## Cleaning Data

### Mengapus nilai NaN dari dataset

```python
df.dropna(how='any', inplace=True)
```

### Mengecek jumlah nilai NaN setelah penghapusan nilai NaN

```python
print(df.isnull().sum())
```

## Data Analysis and Visualization

### Mengetahui harga tertinggi

```python
df1 = df.sort_values(by='price', ascending=False)
df1.head()
```

### Mengetahui harga terendah

### Perbandingan tipe ruangan AirBnb

```python
# Membuat dataframe untuk menghitung jumlah tipe ruangan
room_count_df = df['room_type'].value_counts().reset_index()
room_count_df.columns = ['room_type', 'count']

# Membuat pie chart dengan df.plot()
plt.figure(figsize=(8, 8))
room_count_df.plot(
    kind='pie',
    y='count',
    labels=room_count_df['room_type'],
    autopct='%.2f%%',
    legend=False,
    colors=['#ff9999','#66b3ff','#99ff99'],
    textprops={'fontsize': 10},
    startangle=90,
    shadow=True,
    title='Rasio Tipe Ruangan AirBnb di New York City',
    ylabel=''  # Menghilangkan label y yang tidak perlu
)

plt.tight_layout()
plt.show()
```

### Menghitung jumlah listing AirBnb yang ada di setiap neighbourhood berdasarkan tipe ruangan

```python
neighbourhood_df = df[['neighbourhood_group','neighbourhood','room_type','price','minimum_nights']]

plt.figure(figsize=(10,6))
sns.set(style="whitegrid", color_codes=True)
font_title = {'size': 15}
font_label = {'size': 12}
sns.countplot(
    x='neighbourhood_group',
    data=df,
    hue='room_type',
    palette='dark:yellow'  # Gradien dari dark ke yellow
)
plt.xlabel('Neighbourhood_group',fontdict=font_label,fontweight='bold')
plt.ylabel('Count',fontdict=font_label,fontweight='bold')
plt.title("Jumlah listing AirBnb yang ada di setiap neighbourhood berdasarkan tipe ruangan ",fontdict=font_title,fontweight='bold')
plt.show()
```

### Mengetahui koordinat lokasi seluruh listing AirBnb yang ada di New York City

```python
lokasi_airbnb = df.sort_values(by='price',ascending=False)

#Plotting Scatterplot
plt.figure(figsize=(10,6))
sns.set(style="whitegrid", color_codes=True)
font_title = {'size': 15}
font_label = {'size': 12}
sns.scatterplot(data=lokasi_airbnb,x='longitude',y='latitude',hue='neighbourhood_group')
plt.xlabel('Longitude',fontdict=font_label,fontweight='bold')
plt.ylabel('Latitude',fontdict=font_label,fontweight='bold')
plt.title("Lokasi Seluruh listing AirBnb yang ada di kota New York City",fontweight='bold',fontdict=font_title)
plt.show()
```

### Mengetahui lokasi 1000 listing AirBnb dengan harga tertinggi di kota New York City

```python
lokasi_airbnb = df.sort_values(by='price',ascending=False)[:1000]

#Plotting Scatterplot
plt.figure(figsize=(10,6))
sns.set(style="whitegrid", color_codes=True)
font_title = {'size': 15}
font_label = {'size': 12}
sns.scatterplot(data=lokasi_airbnb,x='longitude',y='latitude',hue='neighbourhood_group')
plt.xlabel('Longitude',fontdict=font_label,fontweight='bold')
plt.ylabel('Latitude',fontdict=font_label,fontweight='bold')
plt.title("1000 Lokasi listing AirBnb dengan harga tertinggi di kota New York City",fontweight='bold',fontdict=font_title)
plt.show()
```

### Mengetahui variasi harga listing berdasarkan tipe ruangan di setiap neighbourhood

```python
plt.figure(figsize=(10,6))
sns.set(style="whitegrid", color_codes=True)
font_title = {'size': 15}
font_label = {'size': 12}
sns.barplot(x = 'neighbourhood_group',y = 'price',hue='room_type',palette = "BuGn_r",data = neighbourhood_df)
plt.xlabel('Neighbourhood_group',fontdict=font_label,fontweight='bold')
plt.ylabel('Price',fontdict=font_label,fontweight='bold')

plt.title("Variasi harga berdasarkan tipe ruangan di setiap neighbourhood",fontweight='bold',fontdict=font_title)
plt.show()
```

### Mengetahui rentang harga di setiap neighbourhood

```python
# Membuat dataframe rentang harga berdasarkan lokasi AirBnb
Queens_Price_df=neighbourhood_df.loc[neighbourhood_df['neighbourhood_group']=='Queens',['neighbourhood_group','price']]
Manhattan_Price_df=neighbourhood_df.loc[neighbourhood_df['neighbourhood_group']=='Manhattan',['neighbourhood_group','price']]
Brooklyn_Price_df=neighbourhood_df.loc[neighbourhood_df['neighbourhood_group']=='Brooklyn',['neighbourhood_group','price']]
Bronx_Price_df=neighbourhood_df.loc[neighbourhood_df['neighbourhood_group']=='Bronx',['neighbourhood_group','price']]
Staten_Island_Price_df=neighbourhood_df.loc[neighbourhood_df['neighbourhood_group']=='Staten Island',['neighbourhood_group','price']]
```

```python
fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(18, 15))
fig.suptitle("Rentang Harga per Wilayah NYC (Log Scale)", fontsize=20, fontweight='bold')

# Data dan judul per subplot
datasets = [
    (Queens_Price_df, "Queens"),
    (Brooklyn_Price_df, "Brooklyn"),
    (Bronx_Price_df, "Bronx"),
    (Staten_Island_Price_df, "Staten Island"),
    (Manhattan_Price_df, "Manhattan"),
]

font_label = {'fontsize': 12}
font_title = {'fontsize': 14}

# Plot tiap wilayah
for i, (data, title) in enumerate(datasets):
    row, col = divmod(i, 2)
    ax = axes[row, col]
    sns.histplot(data=data, x="price", log_scale=True, color='green', ax=ax)
    ax.set_title(f"Rentang harga di {title}", fontdict=font_title, fontweight='bold')
    ax.set_xlabel("Price", fontdict=font_label, fontweight='bold')
    ax.set_ylabel("Count", fontdict=font_label, fontweight='bold')

axes[2, 1].axis('off')

plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```

### Menghitung harga rata-rata harga di setiap neighbourhood

```python
avg_price = df.groupby('neighbourhood_group')['price'].mean().reset_index()
avg_price

# Bar chart rata-rata harga
plt.figure(figsize=(8, 5))
sns.barplot(
    data=avg_price,
    x='neighbourhood_group',
    y='price',
    hue='neighbourhood_group',
    palette='viridis',
    legend=False
)

plt.title('Harga Rata-Rata Airbnb per Neighbourhood Group')
plt.xlabel('Neighbourhood Group')
plt.ylabel('Harga Rata-Rata')
plt.show()
```

# Laporan Singkat

 1. Manhattan: Pusat Akomodasi Mewah

Manhattan jadi wilayah paling mahal di platform Airbnb NYC, dengan harga rata-rata dua kali lipat dibanding Bronx atau Queens. Banyak listing berupa apartemen utuh yang disewakan untuk turis — bukan tempat tinggal biasa. Ini menunjukkan bahwa Manhattan bukan cuma pusat kota, tapi juga pusat bisnis properti digital.

 2. Listing Murah Lebih Populer

Data review menunjukkan listing dengan harga terjangkau di Brooklyn dan Queens mendapat lebih banyak ulasan. Artinya, listing murah cenderung lebih sering disewa, sehingga lebih terlihat di platform. Sementara listing mahal, meskipun mewah, lebih jarang muncul kalau tidak aktif dipromosikan.

 3. Dari Sharing ke Bisnis Properti

Ada sebagian kecil host yang punya puluhan hingga ratusan listing. Di Brooklyn, apartemen utuh bahkan lebih banyak dari kamar pribadi. Ini menandakan banyak orang beli properti khusus untuk disewakan lewat Airbnb — bukan sekadar menyewakan kamar kosong di rumah sendiri.

 4. Harga Favorit di Tengah-Tengah

Sebagian besar booking terjadi di harga 50 - 150 dollar per malam. Manhattan punya dua kelompok harga: satu sekitar 100 dollar (buat budget traveler) dan satu lagi di atas 250 dollar (untuk wisatawan kelas atas). Di Bronx, listing justru tersedia hampir sepanjang tahun — artinya mereka mengandalkan banyak penyewa, bukan harga tinggi.

 5. Lokasi Tetap Jadi Penentu

Model prediksi menunjukkan: semakin banyak properti dimiliki host, harga per listing justru cenderung turun (mungkin karena diskon atau strategi massal). Tapi yang paling berpengaruh tetap lokasi — wilayah seperti Manhattan dan Brooklyn selalu jadi faktor utama penentu harga.

-Investor: Fokus ke properti kecil di area ramai Brooklyn, dengan desain menarik dan harga menengah.
